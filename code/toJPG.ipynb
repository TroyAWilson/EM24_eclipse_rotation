My goal here is to take all the files in the out_new.csv file and make them all jpgs(or pngs) and then create a final csv that is just name of jpg and angle

```python
import csv
import rawpy
from PIL import Image
```

```python
def reset():
    '''
        Wipe file and reset the headers
    '''
    with open('out_final.csv', mode='w') as file:
        csv_writer = csv.writer(file)
        csv_writer.writerow(['FileName', 'Angle'])
        
def addToCSV(file_name, angle):
    '''
        Add row of information to the CSV file
    '''
    with open('out_final.csv', mode='a', newline='') as file:
        csv_writer = csv.writer(file)
        csv_writer.writerow([file_name, angle])
        
# reset()
```

```python
def createJPG(user, file, angle):
    full_path = f'../24_images/{user}/{file}'
    file_no_ext = file.split('.')[0]
    dest_path = f'JPGs/{user}_{file_no_ext}.jpg'
    
    try:
        with rawpy.imread(full_path) as raw:
            rgb = raw.postprocess()
            img = Image.fromarray(rgb)
            img.save(dest_path, "JPEG")
            addToCSV(f'{user}_{file_no_ext}.jpg', angle)
            print(f'{user}_{file_no_ext}.jpg was added to the csv')
    except Exception as e: 
        print(f'There was an issue with {full_path}, error below.\nERROR:{e}')
```

```python
with open('out_new.csv', mode='r') as file:
    csv_reader = csv.reader(file)
    
    for row in csv_reader:
        if row[0] == 'Folder':
            continue
        createJPG(row[0], row[1], row[2])
```
